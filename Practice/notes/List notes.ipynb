##Lists

```python
list1 = [1,2,3,4,5]
print(list1);
```

prints list1 like a list including the braces and commas.

```python
for i in list1:
    print(i, end=" ");
```

prints only the elements in the list

```python
list1.append(6);
print();
print(list1);
```

adds only one element to the list, even if we send a list as argument to the list it adds the list as the element to the list

```python
list1.append([7,8]);
print(list1);
```

The list passed as argument to the append function is added to the original list as a single element. The list elements of the arguments list can to added to the original list by using extend function.

```python
print(list1.pop())
```

```python
print(list1.pop());
list1.extend([6,8,7]);
print(list1);
```

removes the last element of the list and returns that value

extend function adds all the elements from the list passed as arguments to extend function, then directly to out mail list.

```python
list1.sort();
print(list1);
```

sort all the elements according to the values of the list.

```python
print(list1.index(6));
```

The index function returns the index value of the element passed to the function as argument.

```python
list1.append(5);
print(list1.count(5));
```

returns number of elements of same value present in the original list
